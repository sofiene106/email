```python
def check_email(code):
    if ' ' in str(code):
        return np.nan
    if "@" not in str(code):
         return np.nan
    name, domain = str(code).split('@', 1)
    if '.' not in domain[1:]:
        return np.nan
    return str(code)
```

```python
df['EMAIL']=df.apply(lambda x: check_email(x.EMAIL), axis=1)
```

```python
def check_tel(code):
    if ' ' in str(code):
        return np.nan
    if str(code)[0] not in {'2','3','4','5','7','9'} :
        return np.nan
    if len(code) !=8 :
        return np.nan
    return str(code)
```

```python
df['TELEX']=df.apply(lambda x: check_tel(x.TELEX), axis=1)
```
